# Descriptive Statistics

Statistik deskriptif melibatkan peringkasan dan pengorganisasian data sehingga mudah dipahami. Statsitik deskriptif menggambarkan data dalam sampel.

Analisis statistik deskriptif membantu untuk memahami data dan bagian penting dari machine learning. Ini karena pembelajaran mesin adalah tentang membuat prediksi. Di sisi lain, statistik adalah semua tentang menarik kesimpulan dari data, yang merupakan langkah awal yang diperlukan.

### Types of Descriptive Statistics

Dipecah menjadi 2 kategori :

- Ukuran Tendensi sentral
- Ukuran variabilitas(Penyebaran)

### Population and Sample

Populasi adalah sekumpulan semua elemen atau item. Populasi biasanya luas oleh karena itu sulit untuk mengumpulkan dan menganalisis sebuah populasi. Oleh karena itu ahli statistik mencoba membuat beberapa kesimpulan tentang suatu populasi dengan memilih dan memeriksa bagian yang representatif dari populasi bagian tersebug.

Subset dari populasi disebut sampel, idealnya sampel harus memertahankan fitur statistik. Untuk itu kita dapt menggunakan sampel untuk mendapatkan kesimpulan tentang populasi.

### Outliers


Outliers adalah titik data yang berbeda secara signifikan dari mayoritas data yang diambil dari sampel / populasi.

Kesalahan pengumpulan data merupakan penyebab utama outliers. Misal, keterbatasan instrumen / prosedur pengukuran dapat berarti bahwa data yang benar tidak dapat diperoleh. Kesalahan lainnya dapat disebabkan oleh kesalahan peritungan, kontaminasi data, kesalahan manusian dan banyak lagi.



## Measure Of Central Tendency

Kecenderungan sentral mengacu pada gagasan bahwa ada satu angka yang paling baik merangkum seluruh rangkaian pengkuruan, yaitu angka yang ada dalam beberapa cara 'sentra' ke himpunan

### Mean / Average

adalah kecenderungan sentral dari data yaitu angka di mana seluruh data tersebar. Rata-rata dianggap sebagai ukuran yang paling dapat diandalkan dari kecenderungan sentral untuk membuat asumsi tentang populasi dari sampel tunggal



### Median

Median adalah nilai yang membagi data dalam 2 bagian yang sama yaitu jumlah sisi kanan sama dengan jumlah sisi kiri

### Mode

Mode adalah istilah yang muncul waktu maksimum dalam kumpulan data yang memiliki frekuensi tertinggi. Mode adalah nilai atau kategori yang paling sering terjadi dalam data.

## Measure Of Spread / Dispersion

Measure of spread merupakah ukuran penyebaran yang mengacu pada gagasan varibilitas data.

### Standard Deviation

Standar deviasi / simpangan baku adalah pengukuran jarak rata-rata antara setiap kuantitas dan rata-rata. Yaitu bagaimana data disebarkan dari mean. Standar deviasi yang rendah menunjukkan bahwa titik data cenderung dekat dengan rata-rata kumpulan data sedangkan standar deviasi yang tinggi menujukkan bahwa titik data tersebar pada rentang yang lebih luas.


### Variace

Variance adalah rata-rata jarak kuadrat antara setiap kuantitas dan rata-rata. Variance dihitung dengan menemukan perbedaan antara setiap titik data dan rata-rata,mengkuadratkannya,menjumlahkannya dan kemudian mengambil rata-rata angka tersebut.

### Range

Rentang adalah salah satu teknik deskriptif statistik yang paling sederhana. Rentang adalah perbedaan antara nilai terendah dan nilai tertinggi

### Percentile

Persentil adalah cara untuk mewakili posisi suatu nilai dalam kumpulan data. untuk menghitung persentil, nilai dalam kumpulan data harus selalu dalam urutan menaik

### Skewness

Skewness adalah ukuran dari asimetri distribusi probabilitas dari variable acak bernilai riil tentang nilai rata-ratanya. Nilai skewness bisa positif,negatif atau tidak terdefinisi


## Kurtosis

Kurtosis adalah ukuran apakah data memiliki banyak outlier atau sedikit outlier sehingga dapat menunjukkan ukuran keruncingan dan relatif terhadap distribusi normal.






Ada tiga jenis Kurtosis:

- Mesokurtik \ Mesokurtik adalah distribusi yang memiliki kurtosis yang sama dengan kurtosis distribusi normal, yaitu nol.

- Distribusi Leptokurtik \ adalah distribusi yang memiliki kurtosis lebih besar daripada distribusi Mesokurtik. Ekor distribusi seperti itu tebal dan berat. Jika kurva distribusi lebih memuncak dari kurva Mesokurtik, itu disebut sebagai kurva Leptokurtik.

- Distribusi Platykurtic \ adalah distribusi yang memiliki kurtosis lebih rendah dari distribusi Mesokurtic. Ekor distribusi semacam itu lebih tipis. Jika kurva distribusi kurang memuncak dari kurva Mesokurtik, itu disebut sebagai kurva Platykurtic.

Perbedaan utama antara skewness dan kurtosis adalah bahwa skewness mengacu pada tingkat simetri, sedangkan kurtosis mengacu pada tingkat kehadiran pencilan dalam distribusi.

## Correlation

Korelasi adalah teknik statistik yang dapat menunjukkan apakah dan seberapa kuat pasangan variabel saling terkait.


Hasil utama dari suatu korelasi disebut koefisien korelasi atau "r". itu berkisar -1.0 sampai +1.0 semakin dekat r ke angka +1 atau -1 makan semakin dekat kedua variable berkaitan.

Jika r mendekati 0, itu berarti tidak ada hubungan antara variabel. 

Jika r positif, itu berarti bahwa ketika satu variable semakin besar yang lain semakin besar.

Jika r negatif itu berarti ketika satu variabel semakin besar makan yang lain semakin kecil

## Weighted Mean

Weighted mean adalah generalisasi dari rata-rata aritmatika untuk menentukan kontribusi relatif dari setiap titik data terhadap hasilnya.

Weighted mean berguna ketika membutuhkan rata-rata dataset yang berisi item yang terjadi dengan frekuensi relatif yang diberikan.

Misal, kita memiliki set di mana 20% dari semua item sama dengan 2, 50% dari semua item sama dengan 4, dan 30% sisanya dari item sama dengan.

kita dapat menghitung mean seperti set seperti ini :

```python
0.2 * 2 + 0.5 * 4 + 0.3 * 8
```

Di sini, kita dapat memperhitungkan frekuensi dengan bobot. dengan metode ini, kita tida perlu mengetahui jumlah total item.

Kita dapat menerapkan weighted mean dalam python dengan menggabungkan sum() dengan range() atau zip()

```python
x = [8.0, 1, 2.5, 4, 28.0]
w = [0.1, 0.2, 0.3, 0.25, 0.15]

wmean = sum(w[i] * x[i] for i in range (len(x))) / sum(w)
print(wmean)

wmean = sum(x_*w_ for (x_,w_) in zip (x,w)) /sum(w)
print(wmean)
```

cara di atas merupakan implementasi manual dalam python tanpa menggunakan library apapun.

Namun, jika kita memiliki kumpulan data yang besar makan library numpy dapat memberikan solusi yang lebih baik. Kita dapat menggunakan np.average() untuk mendapatkan weighted mean dari sebuah array atau series pandas:

```python
y,z,w = np.array(x), pd.Series(x), np.array(w)

wmean = np.average(y, weights=w)
print(wmean)

wmean=np.average(z, weights=w)
print(wmean)
```

Hasilnya sama seperti dalam kasus implementasi pada python murni.

Cara lain bisa menggunakan syntax seperti di bawah ini :

```python
(w*y).sum() / w.sum()
```

skewness disebut juga sebagai ukuran kemiringan sebuah bilangan yang dapat menunjukkan miring atau tidaknya bentuk kurva suatu distribusi frekuensi.

Skewness sebagai derajat ketidak simetrisan suatu distribusi

```python
#Menggunakan Python

n = len(x)

mean_ = sum(x) / n
var_ = sum((item - mean_)**2 for item in x) / (n-1)
std_ = var_ ** 0.5

skew_ = (sum((item - mean_) **3 for item in x) * n  / ((n-1) * (n - 2) *std_**3))
```
